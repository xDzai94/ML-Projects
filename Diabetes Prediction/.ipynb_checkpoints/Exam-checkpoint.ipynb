# Import library & Config setting

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import shap
import time
import warnings

from matplotlib import pyplot as plt 
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve,accuracy_score,auc,confusion_matrix,roc_auc_score,plot_confusion_matrix, precision_score, recall_score, f1_score, plot_roc_curve
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

warnings.filterwarnings("ignore")
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
```

# Import Data

```python
df = pd.read_csv("diabetes.csv")
display(df.head(), df.info(), df.describe())
```

# EDA & Pre-processing

```python
fig, axis = plt.subplots(3,3,figsize=(20, 8))
df.hist(ax=axis)
```

```python
df.boxplot(figsize=(20, 8))
```

```python
cor_matrix = df.corr().abs()
cor_matrix.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
explode = (0, 0.1)
fig1, ax1 = plt.subplots()
ax1.pie(df['Outcome'].value_counts(),  explode=explode, labels=df['Outcome'].value_counts().index.to_list(), autopct='%1.f%%',
        shadow=True, startangle=90, textprops={'color':"w"})
ax1.axis('equal')
plt.title('Outcome', color='white')
plt.show()
```

## Replace Missing Values 

```python
df['Glucose']=df['Glucose'].replace(0,df['Glucose'].mean())
df['BloodPressure']=df['BloodPressure'].replace(0,df['BloodPressure'].mean())
df['SkinThickness']=df['SkinThickness'].replace(0,df['SkinThickness'].mean())
df['Insulin']=df['Insulin'].replace(0,df['Insulin'].mean())
df['BMI']=df['BMI'].replace(0,df['BMI'].mean())
```

```python
fig, axis = plt.subplots(3,3,figsize=(20, 8))
df.hist(ax=axis)
```

```python
df.describe()
```

## Replace Outliers

```python
columns=['Pregnancies','Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']
                
i = 0 
for cols in columns:
    # Removing Outliers 
    Q3 = df[cols].quantile(0.99)
    df = df[df[cols] <= Q3]
    Q1 = df[cols].quantile(0.01)
    df = df[df[cols] >= Q1]
    i=i+1
```

```python
fig, axis = plt.subplots(3,3,figsize=(20, 8))
df.hist(ax=axis)
```

# Train Test Split

```python
X = df[columns]
y = df['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
X_train, X_val, y_train, y_val =  train_test_split(X_train, y_train, test_size=0.25, random_state=1) # 0.25 x 0.8 = 0.2
```

# Feature Selection

```python
# feature selection
def select_features(X_train, y_train):
    fs = SelectKBest(score_func=mutual_info_classif, k=5)
    fs.fit(X_train, y_train)
    return fs

fs = select_features(X_train, y_train)

sortlist = {}

for i, items in enumerate (X.columns):
    sortlist[items] =  fs.scores_[i]

display(sorted(sortlist.items(), key=lambda x: x[1]))
    
# plot the scores
plt.bar(X.columns, fs.scores_)
plt.xticks(rotation=90)
plt.show()
```

# Finalize & Training

```python
cols = fs.get_support(indices=True)
X = df.iloc[:,cols]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
X_train, X_val, y_train, y_val =  train_test_split(X_train, y_train, test_size=0.25, random_state=1) # 0.25 x 0.8 = 0.2
```

```python
# define scaler
scaler = MinMaxScaler()
scaler.fit(X_train)

# transform all datasets
X_train_scaled = scaler.transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
```

## RF Model

```python
rfc = RandomForestClassifier(n_estimators=10)
rfc.fit(X_train_scaled, y_train)

plot_roc_curve(rfc, X_val_scaled, y_val) 
```

```python
preds = (rfc.predict_proba(X_val_scaled)[:, 1] >= 0.4).astype(int)

acc_rfc = (preds == y_val).sum().astype(float) / len(preds)*100
y_true = y_val
y_pred = preds
rfc_confusion_matrix = confusion_matrix(y_true.tolist(), y_pred.tolist())
rfc_precision_score = precision_score(y_true, y_pred)
rfc_recall_score = recall_score(y_true, y_pred)

print("Scikit-Learn's Random Forest Classifier's prediction accuracy is: %3.2f" % (acc_rfc))
print(f"Confusion matrix:\n {rfc_confusion_matrix}")
print(f"Precision:\n {rfc_precision_score}") 
print(f"Recall:\n {rfc_recall_score}") 
```

## LR Model

```python
lr = LogisticRegression()
lr.fit(X_train_scaled, y_train)

plot_roc_curve(lr, X_val_scaled, y_val) 
```

```python
preds = (lr.predict_proba(X_val_scaled)[:, 1] >= 0.50).astype(int)

acc = (preds == y_val).sum().astype(float) / len(preds)*100
y_true = y_val
y_pred = preds
lr_confusion_matrix = confusion_matrix(y_true, y_pred)
lr_precision_score = precision_score(y_true, y_pred)
lr_recall_score = recall_score(y_true, y_pred)

print("Scikit-Learn's Logistic Regression's prediction accuracy is: %3.2f" % (acc))
print(f"Confusion matrix:\n {lr_confusion_matrix}")
print(f"Precision:\n {lr_precision_score}") 
print(f"Recall:\n {lr_recall_score}") 
```

## LR Selected & Final Tune

```python
preds = (lr.predict_proba(X_val_scaled)[:, 1] >= 0.36).astype(int)

acc = (preds == y_val).sum().astype(float) / len(preds)*100
y_true = y_val
y_pred = preds
lr_confusion_matrix = confusion_matrix(y_true, y_pred)
lr_precision_score = precision_score(y_true, y_pred)
lr_recall_score = recall_score(y_true, y_pred)

print("Scikit-Learn's Logistic Regression's prediction accuracy is: %3.2f" % (acc))
print(f"Confusion matrix:\n {lr_confusion_matrix}")
print(f"Precision:\n {lr_precision_score}") 
print(f"Recall:\n {lr_recall_score}") 
```

## Result Validation with Test Set & Explainable Insight

```python
preds = (lr.predict_proba(X_test_scaled)[:, 1] >= 0.36).astype(int)

acc = (preds == y_test).sum().astype(float) / len(preds)*100
y_true = y_test
y_pred = preds
test_confusion_matrix = confusion_matrix(y_true, y_pred)
test_precision_score = precision_score(y_true, y_pred)
test_recall_score = recall_score(y_true, y_pred)

print("Scikit-Learn's Logistic Regression's prediction accuracy is: %3.2f" % (acc))
print(f"Confusion matrix:\n {test_confusion_matrix}")
print(f"Precision:\n {test_precision_score}") 
print(f"Recall:\n {test_recall_score}") 
```
